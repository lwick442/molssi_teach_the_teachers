```python
import os
```

```python
distance_file = os.path.join("data","distance_data_headers.csv")
```

```python
distances = numpy.genfromtxt(fname=distance_file, delimiter=",", dtype="unicode")
print(distances)
```

```python
headers = distances[0]
print(headers)
```

```python
data = distances[1:]
print(data)
```

```python
# defines new type for every element 
data = data.astype(numpy.float)
print(data)
```

```python
num_columns = len(data[0,:])
print(num_columns)
```

```python
for i in range(1,num_columns):
    column = data[:,i]
    avg_col = numpy.mean(column)
    print(avg_col)
```

```python
for i in range(1,num_columns):
    column = data[:,i]
    avg_col = numpy.mean(column)
    print(F'{headers[i]} {avg_col: .3F}')
    
```

```python
import matplotlib.pyplot
```

```python
# initialize a figure 
matplotlib.pyplot.figure()
matplotlib.pyplot.plot(data[:,1])
```

```python
# rename function 
import matplotlib.pyplot as plt 
plt.figure()
plt.plot(data[:,1])
```

```python
sample = headers[1]

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (Angstrom)')
fig1 = plt.plot(data[:,1:4], label=sample)

plt.legend()

plt.savefig(F'{sample}.png', dpi=300)
```

```python
sample = headers[1:5]

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (Angstrom)')
fig1 = plt.plot(data[:,1:5], label=sample[:])

plt.legend()

plt.savefig(F'{sample}.png', dpi=300)
```

```python
plt.figure()
plt.xlabel("Frame")
plt.ylabel("Distance")
plt.plot(data[:,1], label=headers[1])
plt.plot(data[:,2],label=headers[2])
plt.legend()
```

```python
plt.figure()
plt.xlabel("Frame")
plt.ylabel("Distance")
for col in range(1, len(data[0,:])):
    plt.plot(data[0:100, col], label=headers[col])
plt.legend()
    
                 
```
